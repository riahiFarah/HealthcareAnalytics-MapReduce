```python
import matplotlib.pyplot as plt

# Les données des symptômes et de leurs fréquences
data = [
    ("Alzheimer", "Perte m?moire", 5),
    ("Anemia", "Fatigue", 18),
    ("Anxiety", "Nervosit?", 17),
    ("Asthma", "Essoufflement", 16),
    ("Cancer", "Douleur", 22),
    ("Depression", "Tristesse", 33),
    ("Diabetes", "Soif excessive", 33),
    ("HeartDisease", "Fatigue", 5),
    ("Hypertension", "Vision floue", 36),
    ("Insomnia", "Difficult? ? s'endormir", 20),
    ("KidneyDisease", "Gonflement des jambes et des chevilles", 16),
    ("LiverDisease", "Jauneur de la peau", 6),
    ("Migraine", "Naus?es", 4),
    ("Obesity", "Fatigue", 12),
    ("Thyroid", "Gain ou perte de poids inexpliqu?e", 15)
]

# Extraction des noms de maladies, des symptômes et des fréquences
diseases, symptoms, frequencies = zip(*data)

# Palette de couleurs avec les couleurs spécifiées
color_palette = ['#61A3BA', '#87C4FF', '#D0A2F7', '#ED9ED6', '#7752FE', '#5FBDFF', '#FFC7C7', '#7752FE',
                 '#FF9B82', '#F8FF95', '#A2678A', '#ED7B7B', '#B2A4FF', '#D7BBF5', '#85E6C5']

# Création des graphiques en barres avec la nouvelle palette de couleurs
plt.figure(figsize=(12, 8))
for disease, color in zip(set(diseases), color_palette):
    indices = [i for i, d in enumerate(diseases) if d == disease]
    plt.bar([symptoms[i] for i in indices], [frequencies[i] for i in indices], label=disease, color=color)

plt.xlabel('Symptômes')
plt.ylabel('Fréquence')
plt.title('Fréquence des symptômes pour chaque maladie')
plt.legend()
plt.xticks(rotation='vertical')  # Rotation des étiquettes sur l'axe x
plt.show()
```
